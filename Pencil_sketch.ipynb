We need to read the image in RBG format and then convert it to a grayscale image. 
This will turn an image into a classic black and white photo. 
Then the next thing to do is invert the grayscale image also called negative image, this will be our inverted grayscale image. 
Inversion can be used to enhance details. 
Then we can finally create the pencil sketch by mixing the grayscale image with the inverted blurry image. 
This can be done by dividing the grayscale image by the inverted blurry image. 
Since images are just arrays, we can easily do this programmatically using the divide function from the cv2 library in Python.

Creation of new image(grayscale) by converting the original image to grayscale

```python
## Importing Libraries
import cv2
from PIL import Image
from IPython.display import display
```

```python
## Reading the image and displaying it 

image = cv2.imread("C:/Users/nawin/Downloads/aish.jpg")
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
display(Image.fromarray(image))
```

```python
## Conversion of image into classic black and white image
black_white_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
display(Image.fromarray(black_white_image))

# Observation: Image converted into classic black and white image
```

```python
## Inversion of grayscale image

inverted_image = cv2.bitwise_not(black_white_image)
display(Image.fromarray(inverted_image))

# Observation: Inverted GrayScale image is obtained
```

```python
## creation of Blurred Image

blurred_image = cv2.GaussianBlur(inverted_image, (21, 21), 0)
display(Image.fromarray(blurred_image))

# Observation: Blurred Image is created
```

```python
## Creation the Pencil Sketch 

inverted_blurred = cv2.bitwise_not(blurred_image)
pencil_sketch = cv2.divide(black_white_image, inverted_blurred, scale=250.0)
display(Image.fromarray(pencil_sketch))

## Observation: Pencil Sketch of image is created
```
